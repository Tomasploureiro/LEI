```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import math as math
img = mpimg.imread("polarbear.jpg")
```

```python
def enhanceColor(imgoriginal, fator, canal):
    img = imgoriginal.copy()
    img = img.astype(np.float32)
    img[:,:,canal] = img[:,:,canal] * fator
    img = img.astype(np.uint)
    np.clip(img, 0, 255)
    plt.imshow(img)
    plt.axis("off")
enhanceColor(img, 1.8, 2)
```
